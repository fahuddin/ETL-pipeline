```python
import petl as etl
import pandas as pd
import sqlite3
import glob
import pyodbc
import pymysql
import urllib
import sa
import sys
import os
from sqlalchemy import create_engine
```

```python
class sqlconnection:
    
    def __init__(self,csv):
        self.server = 'medserver.cnmimedicaid.org'
        self.database = 'CMAMasterDB'
        self.username = 'MCUser'
        self.password = 'Medicaid$'
        self.params = urllib.parse.quote_plus( 
            'DRIVER={ODBC Driver 17 for SQL Server};'+
            'SERVER='+self.server+';DATABASE='+self.database+';UID='+self.username+';PWD='+ self.password)
        self.engine = create_engine("mssql+pyodbc:///?odbc_connect=%s" % self.params)
        self.df = pd.read_csv(csv)
    
    def transform_dataframe(self):
        self.df['DOB'] = self.df['DOB'].str.replace('0:00','').replace('12:00:00','')
        self.df['DateService'] = self.df['DateService'].str.replace('0:00','').replace('12:00:00','')
        self.df = self.df.drop(['no'], axis=1)
        self.df=self.df.astype(str)
    
    def getdf(self):
        return self.df
    
    def upload_to_sql(self):
        self.df.to_sql('ClaimsOld',con = self.engine, if_exists = 'append', index=False)
        
        
        
        
        
        
```
